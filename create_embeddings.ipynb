## Notebook to create simple embeddings for the search index

```python
from pathlib import Path
import pandas as pd

csv_path = Path("./Attractions.csv")
df = pd.read_csv(csv_path)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
import numpy as np

all_embeddings = np.vstack([model.encode(str(row[1]["County"])+", "+str(row[1]["Name"])+", "+str(row[1]["Tags"])) for row in df.iterrows()])
```

```python
with Path("./embeddings.npy").open("wb") as f:
    np.save(f, all_embeddings)
```
